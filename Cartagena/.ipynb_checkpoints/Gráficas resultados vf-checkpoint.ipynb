```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
#Read the files:
#---------------
datos = pd.read_csv('pb_45k.csv', delimiter=",", index_col=0)
```

```python
Lista_estados = ['Susceptibles', 'Leves', 'Graves', 'Críticos', 'Recuperados']
Edades = ['0-4', '5-19', '20-39', '40-59', '>60']
```

```python
datos
cont = 0
for col in datos.columns: 
    print(col) 
    cont += 1
```

```python
#Set days:
#---------

from datetime import datetime, timedelta
s="20200302"
print(datos.shape)
x = list(range(datos.shape[0]))
date = datetime(year=int(s[0:4]), month=int(s[4:6]), day=int(s[6:8]))

x_values = [str((date + timedelta(days=d)).strftime("%d/%m/%Y")) for d in x ]
x_values
```

```python
datos["Día"]=x_values
```

```python
datos["Infectados_nuevos_acum"] = 0
for i in range(len(datos["Leves"])):
    if i==0:
        datos["Infectados_nuevos_acum"][i] = datos["Nuevos_infectados"][i]
    else:
        datos["Infectados_nuevos_acum"][i]= datos["Nuevos_infectados"][i]+datos["Infectados_nuevos_acum"][i-1]
datos["Infectados_nuevos_acum"] = datos["Infectados_nuevos_acum"]/datos["Vivos"][0]
```

```python
#Pintando la tasa de ataque del modelo
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
plt.style.use('seaborn-whitegrid')
plt.rcParams["figure.figsize"] = (20,5)

fig = plt.figure()
fig1 = plt.subplot()
No_dias_division = 10

y = datos["Infectados_nuevos_acum"]
x_values = datos["Día"]
fig1.plot(x_values, y)

fig1.set_xticks(x_values[:: No_dias_division])
fig1.set_xticklabels(x_values[::No_dias_division], rotation=45)

plt.ylabel("Tasa de ataque (%)", fontsize='x-large')
plt.xlabel('Date', fontsize="large")

    
plt.legend(loc='upper right', bbox_to_anchor=(1, -0.3), shadow=True)
plt.xticks(fontsize="large")
plt.yticks(fontsize="x-large")
plt.show()
```

```python
#Calculando los conteos porcentualmente
datos["Susceptibles%"]=datos["Susceptibles"]/datos["Vivos"][0]
datos["Leves%"]=datos["Leves"]/datos["Vivos"][0]
datos["Graves%"]=datos["Graves"]/datos["Vivos"][0]
datos["Críticos%"]=datos["Críticos"]/datos["Vivos"][0]
datos["Recuperados%"]=datos["Recuperados"]/datos["Vivos"][0]
```

```python
#Pintando los conteos% de los estados
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
plt.style.use('seaborn-whitegrid')
plt.rcParams["figure.figsize"] = (20,5)

fig = plt.figure()
fig1 = plt.subplot()
No_dias_division = 10

#fig1.plot(datos["Día"],datos["Susceptibles%"],label = "Susceptibles")
fig1.plot(datos["Día"],datos["Leves%"],label = "Leves")
fig1.plot(datos["Día"],datos["Graves%"],label = "Graves")
fig1.plot(datos["Día"],datos["Críticos%"],label = "Críticos")
#fig1.plot(datos["Día"],datos["Recuperados%"],label = "Recuperados")

fig1.set_xticks(x_values[:: No_dias_division])
fig1.set_xticklabels(x_values[::No_dias_division], rotation=45)

plt.ylabel("Conteo (%)", fontsize='x-large')
plt.xlabel('Date', fontsize="large")    
plt.legend(loc='upper right', bbox_to_anchor=(1, -0.3), shadow=True, fontsize = 20)
plt.xticks(fontsize="large")
plt.yticks(fontsize="x-large")
plt.show()
```

```python
#Calculo de las muertes
datos["Muertes_dia"]=0
for i in range(len(datos["Leves"])):
    if i==0:
        datos["Muertes_dia"][i] = 0
    else:
        datos["Muertes_dia"][i]= datos["Vivos"][i-1]-datos["Vivos"][i]
datos["Muertes_acum"]=0
for i in range(len(datos["Leves"])):
    if i==0:
        datos["Muertes_acum"][i] = datos["Muertes_dia"][i]
    else:
        datos["Muertes_acum"][i]= datos["Muertes_acum"][i-1]+datos["Muertes_dia"][i]
```

```python
#pintando las muertes
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
plt.style.use('seaborn-whitegrid')
plt.rcParams["figure.figsize"] = (20,5)

fig = plt.figure()
fig1 = plt.subplot()
No_dias_division = 10

y = datos["Muertes_dia"]
x_values = datos["Día"]
fig1.plot(x_values, y)

fig1.set_xticks(x_values[:: No_dias_division])
fig1.set_xticklabels(x_values[::No_dias_division], rotation=45)

plt.ylabel("Conteo de muertos por día", fontsize='x-large')
plt.xlabel('Date', fontsize="large")
plt.legend(loc='upper right', bbox_to_anchor=(1, -0.3), shadow=True)
plt.xticks(fontsize="large")
plt.yticks(fontsize="x-large")
plt.show()
```
